```python
import numpy as np
import h5py
import matplotlib.pyplot as plt

uniform_bz = h5py.File("uniform_bz.h5")

x  = uniform_bz["x"][:]
vx = uniform_bz["vx"][:]
vy = uniform_bz["vy"][:]
vz = uniform_bz["vz"][:]
```

```python
plt.scatter(vx,vy)
plt.xlabel(r"$V_x$")
plt.ylabel(r"$V_y$")
```

```python
plt.scatter(vx,vz)
plt.xlabel(r"$V_y$")
plt.ylabel(r"$V_z$")
```
