```python
import numpy as np
import h5py
import matplotlib.pyplot as plt

faraday_test = h5py.File("/home/debian/build/tests/faraday/faraday_test.h5")

Bx = faraday_test["Bx"][:]
By = faraday_test["By"][:]
Bz = faraday_test["Bz"][:]

```

```python
x_primal = np.arange(len(Bx)) * 0.02 
x_dual   = np.arange(len(By)) * 0.02 + 0.01  

plt.figure(figsize=(8,5))
plt.plot(x_primal, Bx, label="Bx")
plt.plot(x_dual, By, label="By")
plt.plot(x_dual, Bz, label="Bz")
plt.xlabel("x")
plt.ylabel("B field")
plt.title("Faraday Test")
plt.legend()
plt.show()
```
